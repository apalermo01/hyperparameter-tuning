```python
from hparam_tuning_project.optimization_modules.ray_optimize import run_tuner
import os
import yaml
import json
from ray import tune


def main(cfg):

    search_space = {
        'optimizer_cfg': {
            'args': {
                'lr': tune.loguniform(1e-6, 1e-1)
            }
        }
    }

    results, tuner = run_tuner(cfg=cfg, num_samples=2, num_epochs=2, search_space=search_space, ret_tuner=True)
    return tuner
    # return {
    #     'model_id': cfg['model_cfg']['model_id'],
    #     'split_perc': split,
    #     'best_lrs': best_lrs
    # }

with open(os.path.join("./experiments/basic_lr_run_20241109/training_configs/", "pytorch_classifier_mnist_10.yaml"), "r") as f:
    cfg = yaml.safe_load(f)

tuner = main(cfg)

```

```python

results = {
    'epoch_metrics': tuner.get_results().get_dataframe().to_dict(
        orient='index'),
    'best_metrics': tuner.get_results().get_best_result().metrics,

    'iteration_metrics': {r.metrics['experiment_tag']:
                          r.metrics_dataframe.to_dict(orient='index')
                          for r in tuner.get_results()}
}
```
